```python
# ============================================================
# Extract Orange County from California Counties Shapefile
# ============================================================

import os
import geopandas as gpd


# ----------------------------
# Input / Output Paths
# ----------------------------
INPUT_SHP = r"/Users/ks/Downloads/ca_counties/CA_Counties.shp"
OUTPUT_SHP = r"Orange_County_Outline/Orange_County.shp"

# Name of the county to extract
COUNTY_NAME = "Orange"

# ----------------------------
# Load the counties shapefile
# ----------------------------
counties = gpd.read_file(INPUT_SHP)

print("Columns:")
print(counties.columns.tolist())

# ----------------------------
# Find the county name field
# ----------------------------
possible_fields = ["NAME", "COUNTY", "CountyName", "NAMELSAD", "COUNTY_NAME"]

name_field = None
for field in possible_fields:
    if field in counties.columns:
        name_field = field
        break

if name_field is None:
    raise ValueError(
        "Could not find a county name field.\n"
        f"Available fields:\n{counties.columns.tolist()}"
    )

print(f"Using county name field: {name_field}")

# ----------------------------
# Select Orange County
# ----------------------------
orange = counties[
    counties[name_field]
    .astype(str)
    .str.strip()
    .str.lower()
    == COUNTY_NAME.lower()
    ].copy()

if orange.empty:
    raise ValueError(f"Could not find '{COUNTY_NAME}' in the shapefile.")

# ----------------------------
# Save the new shapefile
# ----------------------------
os.makedirs(os.path.dirname(OUTPUT_SHP), exist_ok = True)

orange.to_file(OUTPUT_SHP)

print(f"Saved Orange County shapefile:")
print(OUTPUT_SHP)

print(f"\nNumber of features: {len(orange)}")
print(f"CRS: {orange.crs}")

orange
```
