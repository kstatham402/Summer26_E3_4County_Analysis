```python
from os import mkdir

import geopandas as gpd

# Read the shapefile
gdf = gpd.read_file("/Users/ks/Desktop/Wu/Summer26/03_CA_Cities/CA_Cities/CA_Cities.shp")

# Check validity
gdf["is_valid"] = gdf.geometry.is_valid

print(f"Total features: {len(gdf)}")
print(f"Invalid features: {(~gdf['is_valid']).sum()}")

# Show which features are invalid
print(gdf.loc[~gdf["is_valid"]])
```

```python
import geopandas as gpd
from shapely import make_valid
import os

# Load shapefile
# gdf = gpd.read_file("your_shapefile.shp")

# Repair invalid geometries
gdf["geometry"] = gdf.geometry.apply(make_valid)

# Verify
print("Invalid after repair:", (~gdf.geometry.is_valid).sum())

# Create output folder
os.makedirs("Fixed_CA_Cities", exist_ok=True)

# Save repaired shapefile
gdf.to_file("Fixed_CA_Cities/Fixed_CA_Cities.shp")

print("Saved repaired shapefile.")
```
